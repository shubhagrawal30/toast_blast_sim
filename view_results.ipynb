```python
import numpy as np
from astropy.io import fits
from matplotlib import pyplot as plt
import os, sys
```

```python
datas, headers = {}, {}
files = os.listdir("./sim/")
for f in files:
    if f.endswith(".fits"):
        print(f)
        with fits.open("./sim/"+f) as hdul:
            data = hdul[0].data
            datas[f] = data
            headers[f] = hdul[0].header
            print(data.shape)
```

```python
for f in datas.keys():
    nz = datas[f].shape[0]
    for i in range(nz):
        if i != 0:
            continue
            pass
        plt.figure()
        plt.imshow(np.log(np.abs(datas[f][i])))
        plt.colorbar()
        plt.title(f"{f} {i}")
        plt.show()
```
